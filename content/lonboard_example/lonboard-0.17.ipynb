```python
import geopandas as gpd
import requests
from io import BytesIO
import lonboard
from zipfile import ZipFile
```

```python
url = "https://cdn.jsdelivr.net/gh/geopandas/geodatasets@main/data_backup/nybb_16a.zip"
r = requests.get(url)
with ZipFile(BytesIO(r.content)) as z:
    z.extractall("/tmp/nybb")
```

```python
gdf = gpd.read_file("/tmp/nybb/nybb_16a/nybb.shp")
```

```python
lonboard.viz(gdf)
```
